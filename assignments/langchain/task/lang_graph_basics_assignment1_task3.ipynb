```python
from typing_extensions import TypedDict,Literal
from langgraph.graph import StateGraph,START,END
```

```python
class UserState(TypedDict):
  user:str
  message:str
  response:str
  is_prime: bool
```

```python
#node 1 function -  welcoming the user
def welcome_user(state:UserState) -> UserState:
  state['message'] += state['user'] + " welcome user!"
  print(state['message'])
  return state

#node2 function - check if user is prime or not

def premium(state: UserState) -> UserState:
    state['response']= " premuim exist for  " + state['user'] + " : free delivery accepted."
    return state

def non_premium(state: UserState) -> UserState:
    state['response'] = "premuim not exist for  " + state['user'] + ": paid delivery."
    return state

# conditional node
def user_check(state: UserState) -> Literal["free", "paid"]:
    if state['is_prime']:
        return "free"
    else:
        return "paid"
```

```python
#intialise graph
system=StateGraph(UserState)
system.add_node("wishes",welcome_user)
system.add_node("free",premium)
system.add_node("paid",non_premium)

system.add_edge(START,"wishes")
system.add_conditional_edges("wishes",user_check,{"free":"free","paid":"paid"})
system.add_edge("free",END)
system.add_edge("paid",END)

graph=system.compile()
```

```python
user = input("hi user state your name: ")
is_prime_input = input("is this user a prime member? (yes/no): ").strip().lower()

initial_state = {
    "user": user,
    "message": "",
    "response": "",
    "is_prime": is_prime_input == "yes"   # converts "yes"/"no" text into a real True/False
}

#Run
result = graph.invoke(initial_state)
print(result["response"])
```
