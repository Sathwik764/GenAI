Build a Student Registration System.

    Check Fees

          ↓

     Fees Paid?
    
      /        \

    Yes         No

       ↓           ↓

    Enroll      Reject

```python
from typing_extensions import TypedDict,Literal   #data validation
from langgraph.graph import StateGraph, START, END  #stategraph will have the graph blueprint we will build-either linear or non linear
import os

#define state
class myState(TypedDict):
    student_name: str
    fees_paid: bool
    status: str
```

```python
def check_fees(state: myState) -> myState:
    print(f"Checking fee status for {state['student_name']}...")
    return state

def enroll(state: myState) -> myState:
    state['status']= state['student_name'] + " fees paid — ENROLLED."
    return state

def reject(state: myState) -> myState:
    state['status'] = state['student_name']  + " fees NOT paid — REJECTED."
    return state

# conditional node
def fee_checker(state: myState) -> Literal["enroll", "reject"]:
    if state['fees_paid']:
        return "enroll"
    else:
        return "reject"
```

```python
#initialise and connect graph
system=StateGraph(myState)
system.add_node("check_fees",check_fees)
system.add_node("fee_paid",enroll)
system.add_node("fee_not_paid",reject)
system.add_edge(START,"check_fees")
system.add_conditional_edges("check_fees",fee_checker,{"enroll":"fee_paid","reject":"fee_not_paid"})
system.add_edge("fee_paid",END)
system.add_edge("fee_not_paid",END)

graph=system.compile()
```

```python

#taking user input
name=(input("enter student name:"))
paid=(input("has the student paid fees? (yes/no):")).strip().lower()

initial_state: myState = myState(
  student_name=name,
  fees_paid=paid=="yes",
  status=""
)

result=graph.invoke(initial_state)
print(result['status'])
```
